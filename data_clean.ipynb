```python
import numpy as np        # Import NumPy for arrays and mathematical operations
import pandas as pd       # Import Pandas for working with data and DataFrames
import seaborn as sns     # Import Seaborn for advanced and visually appealing plots
import matplotlib.pyplot as plt   # Import Matplotlib for basic plotting and data visualization
```

```python
df = pd.read_csv('hotel_bookings.csv')   # Load the dataset from the CSV file into a Pandas DataFrame
```

```python
df = df.drop_duplicates()                        # Remove all duplicate rows from the DataFrame and keep only unique rows
```

```python
df.isna().sum()                                     # Check for missing (NaN) values in each column and count them
```

```python
columns_to_drop = ['reservation_status_date', 'reservation_status', 'company', 'agent']        # List of columns to remove
df = df.drop(columns=columns_to_drop)                                                               # Drop the specified columns from the DataFrame


```

```python
df['country'].fillna(df['country'].mode()[0], inplace=True)  
# Fill missing values in the 'country' column with the most frequent country (mode)

df = df.dropna(subset=["children"])  
# Drop rows where the 'children' column has missing values
```

```python
df.dropna(inplace=True)  
# Remove all remaining rows with any missing values in the DataFrame

```

```python
plt.figure(figsize=(8, 5))  
# Set the size of the plot (width=8, height=5 inches)

sns.histplot(df['adr'], bins=30, kde=True)  
# Create a histogram of the 'adr' column with 30 bins and overlay a KDE (Kernel Density Estimate) curve

plt.title('ADR Distribution')  
# Set the title of the plot

plt.xlabel('ADR (Average Daily Rate)')  
# Label the x-axis

plt.ylabel('Frequency')  
# Label the y-axis

plt.show()  
# Display the plot

```

```python
Q1 = df['adr'].quantile(0.25)  
# Calculate the 25th percentile (first quartile) of the 'adr' column

Q3 = df['adr'].quantile(0.75)  
# Calculate the 75th percentile (third quartile) of the 'adr' column

IQR = Q3 - Q1  
# Calculate the interquartile range (IQR)

lower = Q1 - 1.5 * IQR  
# Define the lower bound for outliers

upper = Q3 + 1.5 * IQR  
# Define the upper bound for outliers

df = df[(df['adr'] >= lower) & (df['adr'] <= upper)]  
# Keep only rows where 'adr' values are within the bounds (remove outliers)

```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['adr'], bins=30, kde=True)
plt.title('ADR Distribution')
plt.xlabel('ADR (Average Daily Rate)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))  
# Set the size of the first plot

sns.boxplot(x=df['adults'])  
# Create a boxplot for the 'adults' column to visualize distribution and outliers

plt.title('Boxplot for Adults')  
# Set the title of the first plot

plt.show()  
# Display the first plot

plt.figure(figsize=(8,4))  
# Set the size of the second plot

sns.boxplot(x=df['stays_in_week_nights'])  
# Create a boxplot for the 'stays_in_week_nights' column to visualize distribution and outliers

plt.title('Boxplot for Stays in Week Nights')  
# Set the title of the second plot

plt.show()  
# Display the second plot
```

```python
df = df[(df['adults'] <= 10) & (df['adults'] >= 1)]  
# Keep only rows where the number of adults is between 1 and 10 (remove unrealistic values)

df = df[(df['stays_in_week_nights'] <= 30) & (df['stays_in_week_nights'] >= 0)]  
# Keep only rows where stays in week nights are between 0 and 30 (remove outliers/unrealistic values)
```

```python
import matplotlib.pyplot as plt  

# Calculate the percentage of bookings without children and without babies
percentages = {
    'No Children': (df['children'] == 0).mean() * 100,  # Percentage of bookings without children
    'No Babies': (df['babies'] == 0).mean() * 100       # Percentage of bookings without babies
}

# Create a bar plot
plt.figure(figsize=(8,5))  
plt.bar(percentages.keys(), percentages.values(), color=['#66b3ff','#99ff99'])  
# Plot the bars with different colors for each category

plt.ylim(0, 100)  
# Set the y-axis limit to 100%

plt.ylabel('Percentage (%)')  
# Label the y-axis

plt.title('Percentage of Bookings Without Children and Babies')  
# Set the title of the plot

# Add the percentage values on top of each bar
for i, v in enumerate(percentages.values()):
    plt.text(i, v + 1, f"{v:.1f}%", ha='center', fontweight='bold')

plt.show()  
# Display the plot

```

```python
# Group the data by hotel type and cancellation status
cancel_by_hotel = df.groupby(['hotel', 'is_canceled']).size().unstack()  
# This creates a table where rows are hotel types and columns are cancellation status (0 or 1)

# Plot a bar chart
cancel_by_hotel.plot(kind='bar', color=['lightgreen', 'lightcoral'])  
# Bar plot showing number of bookings for each hotel type, with different colors for canceled and not canceled

plt.title('Cancellations by Hotel Type', fontsize=14, fontweight='bold')  
# Set the title of the plot with font size and bold weight

plt.xlabel('Hotel Type')  
# Label the x-axis

plt.ylabel('Number of Bookings')  
# Label the y-axis

plt.legend(['Not Canceled', 'Canceled'])  
# Add a legend explaining the colors

plt.xticks(rotation=0)  
# Keep x-axis labels horizontal for better readability

plt.tight_layout()  
# Adjust the layout so everything fits nicely

plt.show()  
# Display the plot

```

```python
# Convert 'arrival_date_month' to a categorical type with a specific order
df['arrival_date_month'] = pd.Categorical(
    df['arrival_date_month'], 
    categories=['January', 'February', 'March', 'April',
                'May', 'June', 'July', 'August',
                'September', 'October', 'November', 'December'],
    ordered=True
)
# This ensures months are sorted chronologically, not alphabetically

# Group data by month and calculate total bookings and average cancellation rate
monthly_data = df.groupby('arrival_date_month').agg({
    'is_canceled': ['count', 'mean']
}).round(3)
# 'count' gives total bookings, 'mean' gives cancellation rate

monthly_data.columns = ['total_bookings', 'cancel_rate']  
# Rename columns for clarity

# Plot total bookings per month
plt.figure(figsize=(12,5))
plt.bar(monthly_data.index, monthly_data['total_bookings'], color='skyblue', edgecolor='black')
plt.title('Monthly Bookings', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
# This bar plot shows the total number of bookings for each month

# Plot monthly cancellation rate
plt.figure(figsize=(12,5))
plt.bar(monthly_data.index, monthly_data['cancel_rate'], color='salmon', edgecolor='black')
plt.title('Monthly Cancellation Rate', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Cancellation Rate')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
# This bar plot shows the average cancellation rate for each month
```

```python
# Calculate the average ADR (Average Daily Rate) per month
monthly_adr = df.groupby('arrival_date_month')['adr'].mean().round(2)
# Groups data by month and computes mean ADR, rounded to 2 decimals

# Plot ADR trend over months
plt.figure(figsize=(12,5))
plt.plot(monthly_adr.index, monthly_adr.values, marker='s', linewidth=2, markersize=8, color='darkgreen')
plt.fill_between(monthly_adr.index, 0, monthly_adr.values, alpha=0.2, color='green')
plt.title('Average Daily Rate (ADR) by Month', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('ADR (â‚¬)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
# Line plot showing how ADR changes month by month with shaded area under the line

# Create a pivot table for average ADR by customer type and hotel type
hotel_customer_adr = pd.pivot_table(
    df, 
    values='adr', 
    index='customer_type', 
    columns='hotel', 
    aggfunc='mean'
).round(2)
# Pivot table with customer types as rows, hotel types as columns, and mean ADR as values

# Plot ADR by hotel type and customer type
hotel_customer_adr.plot(kind='bar', figsize=(12,5), color=['lightblue', 'salmon'])
plt.title('Average ADR by Hotel Type and Customer Type', fontsize=14, fontweight='bold')
plt.xlabel('Customer Type')
plt.ylabel('ADR (â‚¬)')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.legend(title='Hotel Type')
plt.tight_layout()
plt.show()
# Bar plot comparing ADR between different customer types for each hotel type
```

```python
# Count the number of canceled and not canceled bookings
cancel_counts = df['is_canceled'].value_counts()  
# 0 = Not Canceled, 1 = Canceled

# Create a pie chart for booking cancellations
plt.figure(figsize=(6,6))
plt.pie(
    cancel_counts.values, 
    labels=['Not Canceled', 'Canceled'], 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['lightgreen', 'lightcoral']
)
plt.title('Booking Cancellation Rate', fontsize=14, fontweight='bold')
plt.show()
# Pie chart showing the proportion of canceled vs. not canceled bookings
```

```python

# Create lead time categories for bookings
df['lead_time_category'] = pd.cut(
    df['lead_time'], 
    bins=[0, 30, 90, 180, 365, df['lead_time'].max()],
    labels=['0-30 days', '31-90 days', '91-180 days', '181-365 days', '>1 year']
)
# Categorizes 'lead_time' into bins for easier analysis

# Calculate average cancellation rate per lead time category
leadtime_cancel = df.groupby('lead_time_category')['is_canceled'].mean() * 100
# Multiply by 100 to get percentage

# Plot cancellation rate by lead time
plt.figure(figsize=(10,5))
bars = plt.bar(
    leadtime_cancel.index, 
    leadtime_cancel.values,
    color=plt.cm.RdYlGn_r(np.linspace(0, 1, len(leadtime_cancel)))
)
plt.title('Cancellation Rate by Lead Time', fontsize=14, fontweight='bold')
plt.ylabel('Cancellation Rate %')
plt.xlabel('Lead Time')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# Add percentage labels on top of each bar
for bar, value in zip(bars, leadtime_cancel.values):
    plt.text(bar.get_x() + bar.get_width()/2, value + 1, f'{value:.1f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
# Bar plot showing how cancellation rate changes with booking lead time

```

```python

# Select numeric columns for correlation analysis
numeric_cols = [
    'is_canceled', 'lead_time', 'adr', 
    'previous_cancellations', 'days_in_waiting_list', 
    'booking_changes', 'total_of_special_requests', 'is_repeated_guest'
]

# Calculate correlation matrix for the selected numeric columns
corr_matrix = df[numeric_cols].corr()
# Shows the pairwise correlation coefficients between numeric variables

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(
    corr_matrix, 
    annot=True,               # Show correlation values on the heatmap
    cmap='coolwarm',          # Color map for positive/negative correlations
    center=0,                 # Center the colormap at 0
    square=True,              # Make cells square-shaped
    linewidths=1              # Add lines between cells
)
plt.title('Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
# Heatmap visualizing correlations between numeric variables
```

```python
# Analyze average metrics for new vs repeated guests
repeat_guest_analysis = df.groupby('is_repeated_guest')[
    ['is_canceled','adr','total_of_special_requests','lead_time']
].mean().round(2)
# Group by 'is_repeated_guest' and calculate mean of selected columns, rounded to 2 decimals

repeat_guest_analysis.index = ['New Guest','Repeated Guest']  
# Rename the index for better readability

# Plot comparison as a bar chart
repeat_guest_analysis.plot(kind='bar', figsize=(10,5))
plt.title('Comparison between New and Repeated Guests')  
# Set the title of the plot

plt.ylabel('Average Values')  
# Label the y-axis

plt.show()  
# Display the plot
# This bar chart compares average cancellation rate, ADR, special requests, and lead time between new and repeated guests

```

```python
# Calculate total nights stayed during the week and on weekends
week_nights_total = df['stays_in_week_nights'].sum()  
weekend_nights_total = df['stays_in_weekend_nights'].sum()  

# Prepare data for plotting
data = {
    'Type': ['Week Nights', 'Weekend Nights'],
    'Total Nights': [week_nights_total, weekend_nights_total]
}

# Plot total nights stayed as a bar chart
plt.figure(figsize=(8,5))
plt.bar(data['Type'], data['Total Nights'], color=['skyblue','salmon'])
plt.title('Total Nights Stayed: Week vs Weekend')  
# Set the title of the plot

plt.ylabel('Total Nights')  
# Label the y-axis

plt.grid(axis='y', alpha=0.3)  
# Add horizontal grid lines for readability

plt.show()  
# Display the bar chart comparing total week nights vs weekend nights stayed

```
