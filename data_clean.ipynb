```python
columns_to_drop = ['reservation_status_date', 'reservation_status', 'company', 'agent']
data_cleaned = df.drop(columns=columns_to_drop)
```

```python
df['country'].fillna(df['country'].mode()[0], inplace=True)
df = df.dropna(subset=["children"])
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['adr'], bins=30, kde=True)
plt.title('ADR Distribution')
plt.xlabel('ADR (Average Daily Rate)')
plt.ylabel('Frequency')
plt.show()
```

```python
Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['adr'] >= lower) & (df['adr'] <= upper)]
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['adr'], bins=30, kde=True)
plt.title('ADR Distribution')
plt.xlabel('ADR (Average Daily Rate)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['adults'])
plt.title('Boxplot for Adults')
plt.show()

plt.figure(figsize=(8,4))
sns.boxplot(x=df['stays_in_week_nights'])
plt.title('Boxplot for Stays in Week Nights')
plt.show()
```

```python
df = df[(df['adults'] <= 10) & (df['adults'] >= 1)]  
df = df[(df['stays_in_week_nights'] <= 30) & (df['stays_in_week_nights'] >= 0)]
```

```python
cancel_by_hotel = df.groupby(['hotel', 'is_canceled']).size().unstack()

cancel_by_hotel.plot(kind='bar', color=['lightgreen', 'lightcoral'])
plt.title('Cancellations by Hotel Type', fontsize=14, fontweight='bold')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.legend(['Not Canceled', 'Canceled'])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python



df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], 
                                          categories=['January', 'February', 'March', 'April',
                                                     'May', 'June', 'July', 'August',
                                                     'September', 'October', 'November', 'December'],
                                          ordered=True)

monthly_data = df.groupby('arrival_date_month').agg({
    'is_canceled': ['count', 'mean']
}).round(3)
monthly_data.columns = ['total_bookings', 'cancel_rate']

plt.figure(figsize=(12,5))
plt.bar(monthly_data.index, monthly_data['total_bookings'], color='skyblue', edgecolor='black')
plt.title('Monthly Bookings', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()


plt.figure(figsize=(12,5))
plt.bar(monthly_data.index, monthly_data['cancel_rate'], color='salmon', edgecolor='black')
plt.title('Monthly Cancellation Rate', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Cancellation Rate')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python

monthly_adr = df.groupby('arrival_date_month')['adr'].mean().round(2)

plt.figure(figsize=(12,5))
plt.plot(monthly_adr.index, monthly_adr.values, marker='s', linewidth=2, markersize=8, color='darkgreen')
plt.fill_between(monthly_adr.index, 0, monthly_adr.values, alpha=0.2, color='green')
plt.title('Average Daily Rate (ADR) by Month', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('ADR (â‚¬)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

hotel_customer_adr = pd.pivot_table(df, values='adr', 
                                   index='customer_type', 
                                   columns='hotel', 
                                   aggfunc='mean').round(2)

hotel_customer_adr.plot(kind='bar', figsize=(12,5), color=['lightblue', 'salmon'])
plt.title('Average ADR by Hotel Type and Customer Type', fontsize=14, fontweight='bold')
plt.xlabel('Customer Type')
plt.ylabel('ADR (â‚¬)')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.legend(title='Hotel Type')
plt.tight_layout()
plt.show()
```

```python
cancel_counts = df['is_canceled'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(cancel_counts.values, labels=['Not Canceled', 'Canceled'], 
        autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Booking Cancellation Rate', fontsize=14, fontweight='bold')
plt.show()
```

```python

df['lead_time_category'] = pd.cut(df['lead_time'], 
                                 bins=[0, 30, 90, 180, 365, df['lead_time'].max()],
                                 labels=['0-30 days', '31-90 days', '91-180 days', 
                                         '181-365 days', '>1 year'])

leadtime_cancel = df.groupby('lead_time_category')['is_canceled'].mean() * 100

plt.figure(figsize=(10,5))
bars = plt.bar(leadtime_cancel.index, leadtime_cancel.values,
               color=plt.cm.RdYlGn_r(np.linspace(0, 1, len(leadtime_cancel))))
plt.title('Cancellation Rate by Lead Time', fontsize=14, fontweight='bold')
plt.ylabel('Cancellation Rate %')
plt.xlabel('Lead Time')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

for bar, value in zip(bars, leadtime_cancel.values):
    plt.text(bar.get_x() + bar.get_width()/2, value + 1, f'{value:.1f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python

numeric_cols = ['is_canceled', 'lead_time', 'adr', 'total_nights', 
                'total_guests', 'previous_cancellations', 
                'days_in_waiting_list', 'booking_changes',
                'total_of_special_requests', 'is_repeated_guest']

corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, square=True, linewidths=1)
plt.title('Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
repeat_guest_analysis = df.groupby('is_repeated_guest')[['is_canceled','adr','total_nights','total_of_special_requests','lead_time']].mean().round(2)
repeat_guest_analysis.index = ['New Guest','Repeated Guest']
repeat_guest_analysis.plot(kind='bar', figsize=(10,5))
plt.title('Comparison between New and Repeated Guests')
plt.ylabel('Average Values')
plt.show()
```

```python
categorical_cols = [
    'hotel', 'arrival_date_month', 'meal', 'country', 'market_segment', 
    'distribution_channel', 'reserved_room_type', 'assigned_room_type', 
    'deposit_type', 'customer_type', 'reservation_status'
]

for col in categorical_cols:
    counts = df[col].value_counts().head(10)  
    percentages = counts / counts.sum() * 100
    
    plt.figure(figsize=(10,5))
    bars = plt.bar(counts.index, counts.values, color='skyblue', edgecolor='black')
    
    
    for bar, perc in zip(bars, percentages):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50, 
                 f'{perc:.1f}%', ha='center', fontsize=10, fontweight='bold')
    
    plt.title(f'Booking Counts & Percentages by {col}', fontsize=14, fontweight='bold')
    plt.xlabel(col)
    plt.ylabel('Number of Bookings')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.show()
```

Hotel Type

City Hotel (~66%) dominates bookings compared to Resort Hotel (~34%).

Insight: City Hotel is more popular for short-term stays.

Arrival Month

Peak months: August and July; winter months have lower bookings.

Insight: Seasonality is clear â†’ consider off-season promotions.

Meal Plan

Most guests choose BB (Bed & Breakfast) (~77%).

Insight: Guests prefer basic meal options â†’ opportunity to upsell Half Board or Full Board.

Market Segment

Online TA (~47%) and Direct (~20%) dominate.

Insight: High reliance on Online Travel Agents â†’ higher potential cancellations. Encourage direct bookings for better control and profit.

Distribution Channel

TA/TO is the main channel (~82%).

Insight: Traditional agents dominate; direct channel can increase profitability.

Deposit Type

No Deposit (~87%) is most common; others are Refundable or Non-Refundable.

Insight: No-deposit bookings have higher cancellation rates â†’ encourage advance payments.

Customer Type

Most guests are Transient (~75%), with only ~3% being Repeated Guests.

Insight: Guest loyalty is low â†’ implement loyalty programs.
